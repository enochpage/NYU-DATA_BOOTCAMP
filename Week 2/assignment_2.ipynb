## 1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
A = np.random.randint(1, 10, size=(3,4))
B = np.random.randint(1, 20, size=(3,4))
print(A)
print(B)

# stack A and B vertically
AB_vstack = np.vstack((A, B))
print(AB_vstack)

# stack A and B horizontally
AB_hstack = np.hstack((A, B))
print(AB_hstack)
```

## 2. Find common elements between A and B. [Hint: Intersection of two sets]

```python
common = np.intersect1d(A, B)
print(common)
```

## 3. Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]

```python
A[(A >= 5) & (A <= 10)]
# &, | and ~ in numpy mean element-wise logic operators
# The condition produces a Boolean array of the same shape as A.
# When you apply it directly to a 2D array, 
# NumPy flattens the result and returns only the matching elements as a 1D array.
# A[condition] returns flattened elements where the condition is True
```

## 4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])

filtered_rows = iris_2d[(iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5)]
print(filtered_rows)
# The condition (iris_2d[:, 2] > 1.5) & (iris_2d[:, 0] < 5.0) produces a 1D Boolean mask
# one True or False for each row. NumPy selects whole rows where the mask is True.
```

### 7. How to get the rows of a dataframe with row sum > 100?

```python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
# -1 tells NumPy: “figure out the number of rows automatically.”
df.head()
```

```python
df.columns = ['A', 'B', 'C', 'D']
df.head()
```

```python
row_sums = df.sum(axis=1)
df[row_sums > 100]
```
